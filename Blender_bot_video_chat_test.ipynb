```python
from transformers import AutoTokenizer, BlenderbotForConditionalGeneration
mname = "facebook/blenderbot-400M-distill"
model = BlenderbotForConditionalGeneration.from_pretrained(mname)
tokenizer = AutoTokenizer.from_pretrained(mname)

```

```python
def bot_reply(human_msg):
  inputs = tokenizer([human_msg], return_tensors="pt")
  reply_ids = model.generate(**inputs)
  bot_ans=tokenizer.batch_decode(reply_ids, skip_special_tokens=True)[0]
  return bot_ans
```

```python


import datetime
import os 
import librosa

def video_audio_adjust():
  audio_duration=librosa.get_duration(filename="/content/chat_bot/audio/input_audio.wav")
  video_duration=6.000
  print(f"Audio duration= {audio_duration}")
  print(f"Video duration= {video_duration}")
  if audio_duration>video_duration:
      differece=audio_duration-video_duration
      # calculate how many times video should be concat
      times=differece/video_duration
      # create a file with video name and concat the video using ffmpeg
      
      # if times fraction then add 1 to times
      if times%1!=0:
          times+=2
        
      with open("/content/chat_bot/video/concat.txt", "w") as f:
          for i in range(int(times)):
              f.write("file '/content/chat_bot/video/alisa.mp4'")
              f.write("\n")
              
      # concat the video using ffmpeg
      var_1=os.system("ffmpeg -f concat -safe 0 -i /content/chat_bot/video/concat.txt -c copy /content/chat_bot/video/temp.mp4 -y")
      if var_1==0:
        print("video extent")
      
      # cut the new video size of audio duration
      # Write the audio duration in timestamp format
      audio_timestamp = str(datetime.timedelta(seconds=round(audio_duration)+1))
      var_2=os.system(f"ffmpeg -i /content/chat_bot/video/temp.mp4 -ss 00:00:00 -t {audio_timestamp} -async 1 /content/chat_bot/video/input_video.mp4 -y")
      if var_2==0:
        print("video cut")
  else:
      shutil.copy("/content/chat_bot/video/alisa.mp4","/content/chat_bot/video/input_vido.mp4")
      print("video copy")
```

```python
import cv2
def upscale_video():
  if os.path.exists("/content/chat_bot/GFPGAN/inputs/upload"):
    shutil.rmtree("/content/chat_bot/GFPGAN/inputs/upload")
  os.mkdir("/content/chat_bot/GFPGAN/inputs/upload")
  os.chdir("/content/chat_bot/GFPGAN/inputs/upload")
  var5=os.system(f"ffmpeg -i '/content/chat_bot/Wav2Lip/results/result_voice.mp4' %06d.jpg")
  if var5==0:
    print("sync frame extracted")
  os.chdir("/content/chat_bot/GFPGAN")
  if os.path.exists("/content/chat_bot/GFPGAN/results"):
    shutil.rmtree("/content/chat_bot/GFPGAN/results")
  var6=os.system("python inference_gfpgan.py -i /content/chat_bot/GFPGAN/inputs/upload -o results -v 1.3 -s 2 --bg_upsampler realesrgan")
  if var6==0:
    print("upscale")
  
  cap=cv2.VideoCapture('/content/chat_bot/Wav2Lip/results/result_voice.mp4')
  video_fps = cap.get(cv2.CAP_PROP_FPS)
  video_fps=round(video_fps)
  if video_fps==0:
    video_fps=30
  video_name="/content/chat_bot/video/reply.mp4"
  image_list=sorted(os.listdir("/content/chat_bot/GFPGAN/results/restored_imgs/"), key=lambda x: int(x.replace(".jpg","")))
  fourcc = cv2.VideoWriter_fourcc(*'MP4V')
  frame = cv2.imread(f"/content/chat_bot/GFPGAN/results/restored_imgs/{image_list[0]}")
  height, width, layers = frame.shape
  video = cv2.VideoWriter(video_name, fourcc, int(video_fps), (width, height))
  for i in range(len(image_list)):
      # print(i)
      image_path=f"/content/chat_bot/GFPGAN/results/restored_imgs/{image_list[i]}"
      video.write(cv2.imread(image_path))
  video.release()  
  var7=os.system("ffmpeg -i /content/chat_bot/video/reply.mp4 -i /content/chat_bot/audio/input_audio.mp3 -c:v copy -c:a aac -strict experimental -map 0:v:0 -map 1:a:0 /content/chat_bot/video/alisa_reply.mp4")
  if var7==0:
    print("Audio and video syncronized")
  os.system("ffmpeg -i /content/chat_bot/video/alisa_reply.mp4 -vcodec h264 -acodec aac /content/chat_bot/video/alisa_reply_h24.mp4")

  
```
